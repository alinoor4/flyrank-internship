## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
url = "https://raw.githubusercontent.com/alinoor4/flyrank-internship/refs/heads/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

# 1. Compare 'Short' pages vs 'Long' pages
short_pages = df[df['word_count'] < 500]
long_pages = df[df['word_count'] >= 1500]

avg_clicks_short = short_pages['clicks_90d'].mean()
avg_clicks_long = long_pages['clicks_90d'].mean()

print(f"Average clicks for short pages (< 500 words): {avg_clicks_short:.1f}")
print(f"Average clicks for long pages (>= 1500 words): {avg_clicks_long:.1f}")
print(f"Ratio: Long pages get {avg_clicks_long / (avg_clicks_short + 0.1):.1f}x more clicks.\n")
```
